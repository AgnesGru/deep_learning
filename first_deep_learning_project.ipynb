```python
import keras
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = ['the glass of milk', 'the glass of juice', 'the cup of tea', 'I am a good girl', 'I am a good developer',
       'understand the meaning of words', 'your viedeos are good']
```

```python
sent
```

```python
### vocabulary size
voc_size = 10000
```

### One Hot representation

```python
onehot_repr = [one_hot(word, voc_size) for word in sent]
print(onehot_repr)
```

### Word enbeding representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences # all sentences must be the same length
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length = 8
embedded_docs = pad_sequences(onehot_repr, padding = 'pre', maxlen = sent_length)
print(embedded_docs)
```

```python
dim = 10 # how many features
```

```python
model = Sequential()
model.add(Embedding(voc_size, 10, input_length = sent_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```
